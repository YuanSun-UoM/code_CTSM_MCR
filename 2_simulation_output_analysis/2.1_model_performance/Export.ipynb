# Export simulation data over sensors
- This script is used to hourly mean and monthly mean simulation data, respectively.
- Simulations: CNTL_PT, LCZ_PT.

```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
home_path = '/gws/nopw/j04/duicv/yuansun/'
output_dir = home_path + '0_lcz_mcr/output_analysis/single_point/variation_diurnal/'
df_meta = pd.read_csv(home_path + '0_lcz_mcr/output_analysis/single_point/spinup_analysis_period/sensor_metadata.csv')
data_dir = home_path + '0_lcz_mcr/output_analysis/single_point/simulation/sensor_output/'
df_meta
```

```python
hour_average = []
for i, sn in enumerate(df_meta['sensor']):
    df = pd.read_csv(data_dir + sn + '.csv')  
    df['timestamp'] = pd.to_datetime(df['timestamp']) 
    df.set_index('timestamp', inplace=True)
    hourly_avg = df.groupby(df.index.hour).mean()
    hourly_avg['sensor'] = sn
    hour_average.append(hourly_avg)
hour_average_df = pd.concat(hour_average, axis=0).reset_index()
hour_average_df.to_csv(output_dir + 'hourly_mean.csv', index=False)
```

```python
month_average = []
for i, sn in enumerate(df_meta['sensor']):
    df = pd.read_csv(data_dir + sn + '.csv')  
    df['timestamp'] = pd.to_datetime(df['timestamp']) 
    df.set_index('timestamp', inplace=True)
    monthly_avg = df.groupby(df.index.month).mean()
    monthly_avg['sensor'] = sn
    month_average.append(monthly_avg)
month_average_df = pd.concat(month_average, axis=0).reset_index()
month_average_df.to_csv(output_dir + 'monthly_mean.csv', index=False)
```
